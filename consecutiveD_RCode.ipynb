```r
vals<-c(100,50,100,50,100,50,100,50,100,50,
        100,50,100,50,100,50,100,50,100,50)
```

```r
for(value in vals){
    squared_vals<-vals^2
    
    for(squared_step in squared_vals){
        first_squared_step<-squared_vals[1:(length(vals)-1)]
        next_squared_step<-squared_vals[2:(length(vals))]
        numerator<-first_squared_step-next_squared_step     
    }
    
   for(time_step in vals){
       first_step<-vals[1:(length(vals)-1)]
       next_step<-vals[2:(length(vals))]
       denominator<-2*(first_step*next_step)    
    } 
    
    sumPD<-sum(abs(numerator/denominator))
}
consecD<-sumPD*(1/(length(vals)-1))
```
